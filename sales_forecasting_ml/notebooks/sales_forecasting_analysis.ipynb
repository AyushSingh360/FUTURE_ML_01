# Exploratory Data Analysis for Sales Forecasting

In this notebook, we explore the historical sales data to identify trends, seasonality, and patterns across different categories and regions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
# Load data
df = pd.read_csv('../data/processed_data.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

## 1. Historical Sales Trend

```python
plt.figure(figsize=(15, 6))
sns.lineplot(data=df, x='Order Date', y='Sales')
plt.title('Daily Sales Overview (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.savefig('../visualizations/sales_trend.png')
plt.show()
```

## 2. Monthly Seasonality

```python
df['Month'] = df['Order Date'].dt.month
monthly_sales = df.groupby('Month')['Sales'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=monthly_sales, x='Month', y='Sales', palette='viridis')
plt.title('Average Daily Sales by Month')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.tight_layout()
plt.savefig('../visualizations/seasonal_pattern.png')
plt.show()
```

## 3. Weekly Seasonality

```python
df['DayOfWeek'] = df['Order Date'].dt.day_name()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = df.groupby('DayOfWeek')['Sales'].mean().reindex(days_order).reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=weekly_sales, x='DayOfWeek', y='Sales', palette='coolwarm')
plt.title('Average Daily Sales by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Sales')
plt.tight_layout()
plt.show()
```
